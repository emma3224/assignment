## Task 4

**Instructions**

Get daily death cases worldwide.

Note: We are after the cumulative worldwide confirmed cases (summarising daily death cases in overall countries). You should produce ONE series that lists the cumulative worldwide confirmed cases against date, i.e. the index is the date and the value is the cumulative worldwide confirmed cases.

```python
import os
import pandas as pd
from pandas import Series, DataFrame
import numpy as np

url = "https://raw.githubusercontent.com/emma3224/assignment/refs/heads/main/deaths_global_covid19copy101.csv"
data = pd.read_csv(url) # read csv file

date_columns = data.columns[4:]  # date_columns = all columns from index 4 onwards, (thus excludes: provience/state, region/country, longitute, latitute)

cumulative_worldwide = data[date_columns].sum() #adds the value of each column, providing total daily deaths worldwide

cumulative_worldwide.index = pd.to_datetime(cumulative_worldwide.index,format= '%d-%b-%y')  # ensures all dates will be formatted by day, month(abbreviated) and year

monthly = cumulative_worldwide.groupby(cumulative_worldwide.index.to_period('M')).sum() #the sum of all deaths(all countries) and is grouped by month.

monthly.index = monthly.index.to_timestamp() #uses months to create index.

def format_val(x):
    if pd.isna(x): return ' '
    elif x >= 1_000_000: return f'{x/1_000_000:.2f}M' #adding to readability of text, rounding numbers to 2 decimals and if equal to a million adding a suffix 'M' to signify this.
    elif x >= 1_000: return f'{x/1_000:.2f}K' #suffix 'k' to show thousands
    return f'{x:.0f}'


formatted_monthly = monthly.map(format_val) #adds format of numbers to our series
pd.set_option('display.max_rows', None)  # Show all rows

formatted_monthly #displays series


```

## Task 5

**Instructions**

Get daily increase of death cases via defining a function. Hint: Use the death cases of today minus the death cases of yesterday from the data obtained in Task 4.

Note: This time you are calculating  the daily cases worldwide, that is, the NEW cases confirmed just on a specific day and NOT the cumulative number for each day.

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/emma3224/assignment/refs/heads/main/deaths_global_covid19copy101.csv"
data = pd.read_csv(url) #read csv

def get_daily_death_increase(data): #define a function

    date_cols = data.columns[4:] #columns with dates, starting at index 4
    global_total = data[date_cols].sum(axis=0) #global total, sums down each column(all countries) and provides daily deaths
    daily_increase = global_total.diff().fillna(0).round(2) #daily increase - the difference between yesterday and today
    return daily_increase #fillna - fills any non number with 0, .round(2) rounding to 2 decimal places

daily_increase = get_daily_death_increase(data) #operates function

percentage = daily_increase.pct_change() #make the daily increase value a percentage for ease of readability

daily_difference_percentage_change = percentage.fillna(0).round(2).replace([np.inf, -np.inf], 0) #fill non-numbers with 0, round to 2 decimal places and replace and refers to use positive and negative infintive integers if necessary

daily_difference_percentage_change #displays data series
```

## Task 6

**Instructions**

Visualise the data obtained in Task 4 with library matplotlib.

Choose a diagram/plot that makes sense. Please note that you need to plot the data in Task 4, NOT Task 5. Dont forget about all the elements in a plot: Have you labelled your x-axis? How about y-axis? Legend?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.path as path
from matplotlib.ticker import FuncFormatter

url = "https://raw.githubusercontent.com/emma3224/assignment/refs/heads/main/deaths_global_covid19copy101.csv"
data = pd.read_csv(url) #read csv file

date_columns = data.columns[4:]  # all columns excluding non-date values
worldwide = data[date_columns].sum()  # sum of values within date columns (worldwide deaths)
worldwide.index = pd.to_datetime(worldwide.index,format= '%d-%b-%y') #format all dates to day, month(abbreviated), year

def convert_str(x): # define function, convert values if it's a string, to a numerical float value
    if isinstance(x, str):
        x = x.upper().replace(',', '')
        if 'M' in x:
            return float(x.replace('M', '')) * 1_000_000 #add suffix 'M' for millions, ease of reading table
        elif 'K' in x:
            return float(x.replace('K', '')) * 1_000 #add suffix 'K' for thousands
        else:
            return float(x) # returns numerical value with suffix
    return x

cumulative_worldwide = worldwide.apply(convert_str) #applys function

cumulative_worldwide.index = pd.to_datetime(cumulative_worldwide.index)
cumulative_worldwide = cumulative_worldwide.sort_index() #sort index in dates (ascending)

quarterly_sum = cumulative_worldwide.resample('QE').sum() #create sum of daily worldwide deaths into deaths per end of yearly quarters
#chose to do this to make for an easier read of the plot, fit data with certain plot size
plt.figure(figsize=(16,10)) #16 inches by 10 inches plot size

plt.plot(quarterly_sum.index, quarterly_sum.values, marker='o',ms=10, linestyle='-') #create plot, select index, values and choose marker point and size which is a circle.
#select linestyle '-' continuous whole line
quarter_labels = (quarterly_sum.index - pd.offsets.QuarterEnd(startingMonth=3) + pd.offsets.QuarterBegin()).strftime('%b-%Y') #create quarterly labels for x axis
#starts at each 3 months, format month(abbreviated) year.
plt.title('Worldwide Covid-19 Deaths Over Time (Quarterly)',fontsize=25.0, fontweight= 'bold', pad=25) #create title, add space to header area using pad=.

plt.xlabel('Yearly Quarters', fontsize=20,labelpad=25) #formatting x axis label
plt.ylabel('Number of Deaths',fontsize=20,labelpad=25) #formatting y axis label, labelpad (adds space to section)

plt.xticks(ticks=quarterly_sum.index, labels=quarter_labels, rotation=45, fontsize=12) #creates x-axis tick markers
#rotation 45, ensures no overlap

ticks = np.arange(5_000_000, 100_000_001, 5_000_000) #start y axis at 5 Million, end at 10 Million and go up in 5 millions each time. Based on value of Quarterly Sums.
def millions_formatter(x, pos): #define function - custom formatter
    return f'{x/1_000_000:.0f}M' #add suffix 'M' for reader's ease

plt.gca().yaxis.set_major_formatter(FuncFormatter(millions_formatter)) #sets Y-axis to use customer formatter, as defined earlier.

plt.style.use('seaborn-v0_8-pastel') #uses set template plot style
plt.show() #shows plot

```
